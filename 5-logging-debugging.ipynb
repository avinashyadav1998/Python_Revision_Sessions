### Note
if you want to log any error then first apply exception handling then it will capture all errors owr warnings in logging

```python
#e.g

def test_div(a,b):
    
    try:
        lg.info(str(a)+','+str(b))
        r = a/b
        return r
    except Exception as e:
        print("Please checklogs for more info")
        lg.error("Error has occured")
        lg.exception(str(e))      
```

# Some more concepts of logging

```python
import logging as lg2
```

```python
lg2.basicConfig(filename = './logging/advance_LOGS.log',level = lg2.DEBUG,
format='%(asctime)s |:| NAME : %(name)s |:| LEVEL: %(levelname)s |:| MESSAGE: %(message)s')

console_log = lg2.StreamHandler()
console_log.setLevel(lg2.INFO)
datefmt='%Y-%m-%d %H:%M:%S'
console_log.setFormatter(datefmt)
lg2.getLogger().addHandler(console_log)
lg2.info("this is first test for advance logs")
logger1 = lg2.getLogger("logger1.area1")
logger2 = lg2.getLogger("logger2.area2")
```

```python
logger1.info("this is info for logger one")
logger2.debug("this is debug for logger two")
logger2.info("this is  info for logger 2")
```
